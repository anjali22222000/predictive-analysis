# EduPredict — Exploratory Data Analysis
30000 students × 6 months = 180000 records

```python
import pandas as pd, numpy as np, os
BASE=os.path.abspath(os.path.join(os.getcwd(),'..'))
df=pd.read_csv(os.path.join(BASE,'data','student_dataset.csv'))
print('Shape:',df.shape)
df.head()
```

```python
print(df.dtypes)
print(df.isnull().sum())
df.describe().round(2)
```

```python
import matplotlib; matplotlib.use('Agg')
import matplotlib.pyplot as plt
fig,axes=plt.subplots(2,3,figsize=(15,8))
cols=['Exam_Score','Hours_Studied','Attendance','Previous_Scores','Test_Score','Project_Marks']
colors=['#3b82f6','#10b981','#8b5cf6','#f59e0b','#ef4444','#22d3ee']
for ax,col,c in zip(axes.flat,cols,colors):
    ax.hist(df[col],bins=30,color=c,alpha=0.8,edgecolor='white'); ax.set_title(col)
plt.tight_layout(); plt.savefig(os.path.join(BASE,'outputs','graphs','distributions.png'),dpi=150); plt.show()
```

```python
corr=df.select_dtypes(include=[np.number]).corr()
print('Top correlations with Exam_Score:')
print(corr['Exam_Score'].sort_values(ascending=False).round(3))
```

```python
monthly=df.groupby('Month')['Exam_Score'].mean().reindex(['Jan','Feb','Mar','Apr','May','Jun'])
fig,ax=plt.subplots(figsize=(8,4))
ax.plot(monthly.index,monthly.values,marker='o',color='#3b82f6',linewidth=2)
ax.fill_between(monthly.index,monthly.values,alpha=0.15,color='#3b82f6')
ax.set_title('Monthly Avg Score'); plt.tight_layout(); plt.savefig(os.path.join(BASE,'outputs','graphs','monthly.png'),dpi=150); plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,5))
sc=ax.scatter(df['Hours_Studied'],df['Exam_Score'],c=df['Attendance'],cmap='viridis',alpha=0.4,s=12)
plt.colorbar(sc,label='Attendance %'); ax.set_xlabel('Hours Studied'); ax.set_ylabel('Exam Score')
plt.tight_layout(); plt.savefig(os.path.join(BASE,'outputs','graphs','scatter.png'),dpi=150); plt.show()
print('EDA complete!')
```
